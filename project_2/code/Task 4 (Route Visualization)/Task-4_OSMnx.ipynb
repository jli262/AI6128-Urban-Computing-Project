### Task 4 Route Visualization (Method 1: Osmnx)

```python
import osmnx as ox
import pandas as pd
import matplotlib.pyplot as plt
import folium

import folium
import pandas as pd
import geopandas as gpd
import shapely
import shapely.wkt
```

```python
json_path = '/content/drive/MyDrive/Colab Notebooks/AI6128_Project2/porto/porto.geojson'
route_path = '/content/drive/MyDrive/Colab Notebooks/AI6128_Project2/porto/match_result.csv'
output_dir= '/content/drive/MyDrive/Colab Notebooks/AI6128_Project2/output/'
```

```python
# import results trajectory dataset from task 3
import shapely.wkt
df = pd.read_csv(route_path)
df = pd.DataFrame(df)
df = df.loc[df['mgeom'].str.contains("LINESTRING")]
df = df.loc[df['mgeom'] != 'LINESTRING()']
# df = df.dropna(axis=0, how='any', inplace=False)
# df['mgeom'] = df['mgeom'].apply(shapely.wkt.loads)
df = df.reset_index(drop=True)
df.head(10)
```

```python
def single_route_visualization(trip_number, color, network_graph, save_dir=None):
    fig, ax = ox.plot_graph(network_graph, node_size=1, figsize=(15, 15), bgcolor = '#FFFFFF', edge_linewidth=2, show=False)

    mgeom = [[float(j) for j in i.split(' ')] for i in df['mgeom'][trip_number][11:-1].split(',')]
    
    # print(f"mgeom {mgeom}")
    x, y = zip (*mgeom)
    # print(f"x {x} y {y}")
    ax.plot(x, y, c=color, linewidth=6)
    x_max, x_min, x_gap, x_mid = max(x), min(x), max(x) - min(x), (max(x) + min(x))/2
    y_max, y_min, y_gap, y_mid = max(y), min(y), max(y) - min(y), (max(y) + min(y))/2
    gap = max(x_gap, y_gap) * 0.6
    ax.set_xlim(x_mid - gap, x_mid + gap)
    ax.set_ylim(y_mid - gap, y_mid + gap)

    print("")
    plt.title(f"Route Visualization for Trip Number {trip_number}", fontsize=20)
    plt.show()

    if save_dir:
      fig.savefig(save_dir)
```

```python
def multi_route_visualization(network_graph, save_dir=None):
  fig, ax = ox.plot_graph(network_graph, node_size=1, figsize=(15, 30), bgcolor = '#FFFFFF', edge_linewidth=2, show=False)

  x_max, y_max = float('-inf') , float('-inf') 
  x_min, y_min = float('inf'),  float('inf')

  colors = enumerate(('red',  'orange', 'yellow',  'olive', 'green', 'blue', 'purple', 'cyan',  'pink', 'black'))
  for trip_number, color in colors:
    
    mgeom = [[float(j) for j in i.split(' ')] for i in df['mgeom'][trip_number][11:-1].split(',')]
  
    x, y = zip (*mgeom)
    ax.plot(x, y, c=color, linewidth=6)
    x_min = min(min(x), x_min)
    x_max = max(max(x), x_max)
    y_min = min(min(y), y_min)
    y_max = max(max(y), y_max)
  x_gap, y_gap = (x_max - x_min), (y_max - y_min)
  x_mid, y_mid = (x_max + x_min)/2, (y_max + y_min)/2
  gap = max(x_gap, y_gap) * 0.6
  ax.set_xlim(x_mid - gap, x_mid + gap)
  ax.set_ylim(y_mid - gap, y_mid + gap)

  print("")
  plt.title('Route Visualization for First 10 Trips', fontsize=20)
  plt.show()
  if save_dir:
    fig.savefig(save_dir)
```

```python
road_network=ox.graph_from_place('Porto Portugal', network_type='drive')
```

```python
# main
# road_network = road_network.graph
colors = enumerate(('red',  'orange', 'yellow',  'olive', 'green', 'blue', 'purple', 'cyan',  'pink', 'black'))

# Single route visualization
for trip_number, color in colors:
  single_route_visualization(trip_number,color,road_network, save_dir = output_dir + f'Route-{trip_number+1}.png') 

# Multiple route visualixation
multi_route_visualization(road_network, save_dir = output_dir + "Route_Visualization.png")
```
